```python

# 1.A Ekstrak data transaksi dari file CSV
import pandas as pd

from IPython.display import display
df_csv = pd.read_csv("transactions.csv")

df_csv.to_parquet('transactions.parquet', compression='snappy')

display(df_csv.head());
```

```python
# 1.B Ekstrak data produk dari file excel
df_excel = pd.read_excel("products.xlsx")

df_excel.to_parquet('products.parquet', compression='snappy')

display(df_excel.head());
```

```python
# 2. impor data dari database menggunakan pymysql

from pymysql import connect

connection = connect(
    host = 'localhost',
    database = 'mydb',
    user = 'root',
    password = 'hikmal'
)

cursor = connection.cursor()

cursor.execute('SELECT * FROM customers')

results = cursor.fetchall()

connection.close()

df_customers = pd.DataFrame(results, columns=['id', 'name', 'email', 'address'])

df_customers.to_parquet('customers.parquet', compression='snappy')

df_customers.head()
```

```python
# 3. Fetch Data API 
import requests


url = 'https://fakestoreapi.com/products'

df = pd.json_normalize(requests.get(url).json())

df.to_parquet('api_products.parquet', compression='snappy')

df.head()
```
