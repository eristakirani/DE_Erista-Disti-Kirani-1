```python
# Prioritas 2
# Scrapping data dari web e-commerce

import requests
import pandas as pd
from bs4 import BeautifulSoup
from IPython.display import display
import time

def scrape_reviews(url):
    max_attempts = 3
    attempts = 0
    
    while attempts < max_attempts:
        try:
            response = requests.get(url)
            response.raise_for_status()
            html = BeautifulSoup(response.content, 'html.parser')

            reviews_list = []
            reviews = html.find_all("div", class_="col _2wzgFH K0kLPL _1QgsS5")

            for review in reviews:
                name = review.find("p", class_="_2sc7ZR _2V5EHH _1QgsS5").get_text().strip()
                rating = review.find("div", class_="_3LWZlK").get_text().strip()
                comment = review.find("div", class_="_6K-7Co").get_text().strip()
                date = review.find_all("p", class_="_2sc7ZR")[1].get_text().strip()
                reviews_list.append([name, rating, comment, date])

            return reviews_list
        except requests.exceptions.RequestException as e:
            print(f"Error: {e}")
            attempts += 1
            time.sleep(1)

    return []

df_reviews_all = pd.DataFrame(columns=['name', 'rating', 'comment', 'date'])
for i in range(1, 11):
    url = f"https://www.flipkart.com/adidas-step-n-pace-m-running-shoes-men/product-reviews/itm552d6f5d12438?pid=SHOGZT6GRFFXDKJT&lid=LSTSHOGZT6GRFFXDKJT7N7UOQ&marketplace=FLIPKART&page={i}"
    reviews_data = scrape_reviews(url)
    df_reviews_all = pd.concat([df_reviews_all, pd.DataFrame(reviews_data, columns=['name', 'rating', 'comment', 'date'])], ignore_index=True)

display(df_reviews_all)
```

```python
df_reviews_all.to_csv("data-reviews.csv", index=False)
df_reviews_all.to_excel("data-reviews.xlsx", index=False)
```
