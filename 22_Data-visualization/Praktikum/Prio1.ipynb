#### Import Library dan Extract Data

```python
import pandas as pd
import matplotlib.pyplot as plt

json = "https://gist.githubusercontent.com/nadirbslmh/8fc9cc6cd5cbaaf5cbff63b090fb497e/raw/a7bf3e1edab88b04314a40a9de3ed744bc86d0e9/ecommerce.json"
data = pd.read_json(json)
df = pd.DataFrame(data)
df
```

### Soal Nomor. 1

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format='%d-%m-%Y')

plt.figure(figsize=(12, 18))

# Plot 1: Grafik Garis - Total Nilai Transaksi Harian
total_harian = df.groupby(df['transaction_date'].dt.floor('D'))['transaction_amount'].sum()
plt.plot(total_harian, marker='o')
plt.title('Total Nilai Transaksi Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Nilai Transaksi (IDR)')
plt.grid(True)
plt.tight_layout()

# Plot 2: Bar Plot - Jumlah Pelanggan per Hari
plt.figure(figsize=(12, 6))
customer_per_day = df.groupby(df['transaction_date']).size()
customer_per_day.plot(kind='bar', color='orange')
plt.title('Jumlah Pelanggan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pelanggan')
plt.tight_layout()

# Plot 3: Scatter Plot - Distribusi Nilai Transaksi Harian
plt.figure(figsize=(12, 6))
total_harian = df.groupby(df['transaction_date'].dt.floor('D'))['transaction_amount'].sum()
plt.scatter(total_harian.index, total_harian.values, color='blue')
plt.title('Distribusi Nilai Transaksi Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Nilai Transaksi (IDR)')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Soal Nomor. 2

```python
data_bakery = pd.read_csv("bakery.csv")
df_bakery = pd.DataFrame(data_bakery)
df_bakery
```

```python
df_bakery['transaction_date'] = pd.to_datetime(df_bakery['transaction_date'], format='%d-%m-%Y')

daily_sales = df_bakery.groupby(df_bakery['transaction_date'].dt.date)['transaction_amount'].sum().reset_index()

plt.figure(figsize=(10, 5))
plt.bar(daily_sales['transaction_date'], daily_sales['transaction_amount'], color='skyblue')
plt.title('Grafik Total Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan (IDR)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))
plt.barh(df_bakery['item_name'].value_counts().index, df_bakery['item_name'].value_counts().values, color='salmon')
plt.title('Grafik Jenis Roti yang Paling Sering Dipesan')
plt.xlabel('Jumlah Pesanan')
plt.ylabel('Jenis Roti')
plt.show()

average_customer_satisfaction = df_bakery.groupby(df_bakery['transaction_date'].dt.date)['customer_rating'].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(average_customer_satisfaction['transaction_date'], average_customer_satisfaction['customer_rating'], marker='o', color='green')
plt.title('Grafik Rata-Rata Kepuasan Pelanggan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Rata-Rata Kepuasan Pelanggan')
plt.tight_layout()
plt.show()
```
