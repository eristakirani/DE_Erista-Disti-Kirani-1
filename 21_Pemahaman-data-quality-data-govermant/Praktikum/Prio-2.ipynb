### Import Library & Extract Data

```python
import pandas as pd

df = pd.read_csv("tx_data.csv")
df
```

```python
# Mengubah nomor telepon menjadi string dan menambahkan '+62' di depannya
df['phone_number'] = df['phone_number'].apply(lambda x: '+62' + str(x))

# Menampilkan DataFrame yang telah dimodifikasi
df
```

```python
# Mengonversi nilai 'transaction_amount' dan menambahkan awalan mata uang 'Rp '
df['transaction_amount'] = df['transaction_amount'].apply(lambda x: "Rp {:,.0f}".format(x).replace(",", "."))

# Menampilkan DataFrame yang telah dimodifikasi
df
```

```python
df = df[df['transaction_status'] == 'success']
df
```
